```python
import numpy as np
import pandas as pd
import os
```

```python
PREDICTIONS_FOLDER = os.path.join('data', 'predictions')
all_data_predicted = pd.read_csv(os.path.join(PREDICTIONS_FOLDER, 'all_data_predicted.csv'))
all_data_predicted
```

```python
all_data_predicted.columns
```

```python
affective_categories = ['Expression of emotions', 'Use of humor', 'Self-disclosure']
interactive_categories = ['isReply', "Referring explicitly to others' messages", 'Asking questions', 'Complimenting, expressing appreciation or criticism', 'Expressing agreement or disagreement']
cohesive_categories = ['Vocatives', 'Addresses or refers to the group using inclusive pronouns', 'Phatics, salutations']
all_data_predicted['No category'] = (~(all_data_predicted[all_data_predicted.columns[8:]].any(axis='columns') | all_data_predicted['isReply'] == 1)).astype(int)
all_data_predicted['Affective'] = all_data_predicted[affective_categories].any(axis='columns').astype(int)
all_data_predicted['Interactive'] = all_data_predicted[interactive_categories].any(axis='columns').astype(int)
all_data_predicted['Cohesive'] = all_data_predicted[cohesive_categories].any(axis='columns').astype(int)
all_data_predicted
```

```python
def get_results(column):
    positives = all_data_predicted[all_data_predicted[column] == 1]
    print('Total instances for column "{}": {}'.format(column, positives.shape[0]))
    print('Section distribution in positves', np.unique(positives['section'], return_counts=True), '\n')

_ = pd.Series(list(all_data_predicted.columns[8:]) + ['isReply']).apply(get_results)
```
